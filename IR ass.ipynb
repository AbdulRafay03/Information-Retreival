# Preprocesssing

```python
import re
from nltk.stem.porter import PorterStemmer

port_stem = PorterStemmer()
def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ',content) #Removing other characters
    stemmed_content = stemmed_content.lower() # creating tokens
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords] # stemming and removing stopwords
    # stemmed_content = ' '.join(stemmed_content)
    return stemmed_content

temp = []
for i in (articles):
    temp.append(stemming(i))

articles = temp
print(articles)
```

# Creating Vocab and Inverted Indexes

```python
vocab = []
postings = {}


for i in articles:
    for j in i:
        if j not in vocab:
            vocab.append(j)
print(len(vocab))
vocab.sort()

for i in vocab:
    postings[i] = []

# stemming reduced vocab size from 20k to 16k
```

```python

for i in vocab:
    for a in range(len(articles)):
        if i in articles[a] : 
            postings[i].append(a)
    
```

# Query Processing

### inverted index QUERY

```python
l = [1,2,3,7,8,9,11,12,13,14,15,16,17,18,21,22,23,24,25,26]
def stemming2(content): #pre processing qeury
    stemmed_content = re.sub('[^a-zA-Z]',' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) if word not in ['and', 'or', 'not'] 
                       else word for word in stemmed_content]
    # stemmed_content = ' '.join(stemmed_content)
    return stemmed_content

def posting_q(qeury):
    
    qeury = stemming2(qeury)
    result = []
    for i in range(len(qeury)):
        if qeury[i] not in vocab: 
            return 'Null'
        
        if qeury[i] in ['and' , 'or']: #spliting qeury by operators 
            if qeury[i+1] != 'not' :
                if not result: 
                    peechay_wala = postings[qeury[i-1]]
                else: 
                    peechay_wala = result

                aagay_wala = postings[qeury[i+1]] 
            else:
                if not result: 
                    peechay_wala = postings[qeury[i-1]]
                else: 
                    peechay_wala = result

                aagay_wala = process_not(qeury, i+1)
                if qeury[i] == 'and':
                    result =  set(peechay_wala) & set(aagay_wala)
                elif qeury[i] == 'or':
                     result =  set(peechay_wala) | set(aagay_wala)
        elif qeury[i] == 'not':
            result = process_not
        elif ['and' , 'or' , 'not'] not in qeury:
            result =  postings[qeury[0]]
    
    for i in range(len(result)):
        result[i] = l[i] # converting to the orignal names
    return result

def process_not(qeury , i): # processing NOT seprately
    l = [1,2,3,7,8,9,11,12,13,14,15,16,17,18,21,22,23,24,25,26]
    result_list = [item for item in l if item not in postings[qeury[i+1]]]
    return result_list

print(posting_q('F.,&'))
```

# Positional Qeury Processing


```python
def proximity_search(positional_index, term1, term2, max_distance):
    if term1 not in positional_index or term2 not in positional_index:
        return []

    result = []
    for doc in positional_index[term1]:
        if doc in positional_index[term2]:
            positions1 = positional_index[term1][doc]
            positions2 = positional_index[term2][doc]
            for pos1 in positions1:
                for pos2 in positions2:
                    if abs(pos1 - pos2) <= max_distance:
                        result.append((doc, pos1, pos2))
    return result

# Example usage
term1 = "machine"
term2 = "learning "
max_distance = 0
proximity_results = proximity_search(positional, term1, term2, max_distance)
print("Proximity search results:", proximity_results)
```
